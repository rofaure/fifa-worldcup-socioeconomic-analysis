```python
import pandas as pd
from pathlib import Path
```

```python
root_dir = Path.cwd()
if not (root_dir / "data").exists():
    root_dir = root_dir.parent
if not (root_dir / "data").exists():
    raise FileNotFoundError("Could not locate project root containing the data directory.")

bronze_dir = root_dir / "data" / "bronze"
bronze_dir.mkdir(parents=True, exist_ok=True)

gdp_path = bronze_dir / "bronze_gdp.csv"
pop_path = bronze_dir / "bronze_pop.csv"
matches_path = bronze_dir / "bronze_wc_matches.csv"
```

```python
gdp = pd.read_csv(gdp_path)
pop = pd.read_csv(pop_path)
matches = pd.read_csv(matches_path)
```

```python
tables = [pop, gdp, matches[matches['year']>=1998]]
table_names = ['pop', 'gdp', 'matches']
country_column = ['Country Name', 'Country Name', 'home_team']
country_tables = []

for i, name in enumerate(table_names) :
    print(name, 'table country number :', len(tables[i][country_column[i]].unique()))
    country_tables.append(tables[i][country_column[i]].unique())
```
