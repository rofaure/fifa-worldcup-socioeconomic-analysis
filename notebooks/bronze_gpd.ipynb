# Bronze Ingestion for GDP

## Imports

```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import uuid
```

## Paths

```python
root_dir = Path.cwd()
if not (root_dir / "data").exists():
    root_dir = root_dir.parent
if not (root_dir / "data").exists():
    raise FileNotFoundError("Could not locate project root containing the data directory.")

raw_dir = root_dir / "data" / "raw"
bronze_dir = root_dir / "data" / "bronze"
bronze_dir.mkdir(parents=True, exist_ok=True)

source_path = raw_dir / "world_bank_data" / "API_GDP" / "API_NY.GDP.MKTP.CD_DS2_en_csv_v2_446954.csv"
output_path = bronze_dir / "bronze_gdp.csv"
```

## Bronze ingestion

```python
gdp_df = pd.read_csv(
    source_path,
    skiprows=4    
)
gdp_df = gdp_df.copy()
gdp_df["source_name"] = "fifa_world_cup_archive"
gdp_df["load_timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
gdp_df["run_id"] = str(uuid.uuid4())

gdp_df.to_csv(output_path, index=False)
gdp_df.head()
```
