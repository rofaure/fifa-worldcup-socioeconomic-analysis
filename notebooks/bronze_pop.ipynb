```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import uuid
```

```python
root_dir = Path("..")
if not (root_dir / "data").exists():
    root_dir = root_dir.parent
if not (root_dir / "data").exists():
    raise FileNotFoundError("Could not locate project root containing the data directory.")

bronze_dir = root_dir / "data" / "bronze"
bronze_dir.mkdir(parents=True, exist_ok=True)

source_path = bronze_dir / "world_bank_data" / "API_SP.POP.TOTL_DS2_en_csv_v2_450890.csv"
output_path = bronze_dir / "bronze_pop.csv"
```

```python
pop_df = pd.read_csv(
    bronze_dir / "world_bank_data" / "API_POP" / "API_SP.POP.TOTL_DS2_en_csv_v2_450890.csv",
    skiprows=4
    )

pop_df = pop_df.copy()
pop_df["source_name"] = "pop.worldbank"
pop_df["load_timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
pop_df["run_id"] = str(uuid.uuid4())


pop_df.to_csv(output_path, index=False)
pop_df.head()
```
