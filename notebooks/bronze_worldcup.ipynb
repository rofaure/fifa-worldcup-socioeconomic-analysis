# Bronze Ingestion for World Cup matches

## Imports

```python
import pandas as pd
from pathlib import Path
from datetime import datetime
import uuid
```

## Paths

```python
root_dir = Path.cwd()
if not (root_dir / "data").exists():
    root_dir = root_dir.parent
if not (root_dir / "data").exists():
    raise FileNotFoundError("Could not locate project root containing the data directory.")

bronze_dir = root_dir / "data" / "bronze"
bronze_dir.mkdir(parents=True, exist_ok=True)

source_path = bronze_dir / "fifa_world_cup" / "wcmatches.csv"
output_path = bronze_dir / "bronze_wc_matches.csv"
```

# Bronze ingestion

```python
wc_df = pd.read_csv(source_path)

wc_df = wc_df.copy()
wc_df["source_name"] = "fifa_world_cup_archive"
wc_df["load_timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
wc_df["run_id"] = str(uuid.uuid4())

wc_df.to_csv(output_path, index=False)
wc_df.head()
```
